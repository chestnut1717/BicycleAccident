```python
# 위의 문자열들을 각 지역별로 저장하기 위함
def replace_to_list(long_str):
    long_str = long_str.replace(";", ",")
    long_str = long_str.replace("SetArray", "")
    long_str = long_str.replace("\n\t", "")
    long_str = long_str.replace(" ", "")
    tmp = list(eval(long_str))
    
    replaced = []
    
    for tpl in tmp:
        replaced.append( (int(tpl[2]), tpl[3]) )
    return replaced
```

```python
def error_check(tbody_list):
    if tbody_list[0].get_text() == 'null':
        return True
    return False
```

```python
def get_data(soup, sido_name, city_name):
    
    # 순번이 초과되어 null값이 나올 때, 예외처리
    # 중요한 건, 주소값은 null이 나올 수도 있으니, 주소를 기준으로 하면 안 된다
    tbody_list = soup.select('tbody')[1].find_all('td')
    is_error = error_check(tbody_list)
    
    if is_error:
        return False
    
    # 정상적으로 값이 존재하는 경우
    else:
        data_list = []
        
        for i in range(6):
            tmp = soup.select('tbody')[i].find_all('td')

            # tmp list에서 0번째 index에 있는 것이 지역정보
            if i == 0:
                area_data = tmp[0].find_all('dd')
                
                road = area_data[1].get_text()
                
                    
                # join을 하는 이유는, 주소에 읍,동 등이 추가되어 나오기 때문. 나중에 join할 때 문제 생길수도
                road = " ".join(road.split()[1:])
                data_list.append(sido_name)
                data_list.append(city_name)
                data_list.append(road)

            # 나머지는 사고 정보(날씨, 법규위반 등)
            else:
                for t in tmp:
                    num_data = int(t.get_text())
                    data_list.append(num_data)

        return data_list
```
