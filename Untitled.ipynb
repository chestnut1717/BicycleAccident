```python
import pandas as pd

# 한글 단어 빈도분석
# hannanum, kkma Okt(twitter) 중에 kkma의 형태소 분석이 중복 처리 문제도 없고, 가장 세부적으로 분석했다.
from konlpy.tag import Kkma
kkma = Kkma()

import re
```

```python
df = pd.read_excel('./final_data/교통약자다발지점_사고지표_세부정보_병합.xlsx')
df
```

```python
not_capital = df.loc[(df['광역'] != '서울') & (df['광역'] != '경기'), ['광역','지점명']]
not_capital
```

```python
not_capital['지점명'].values
```

```python
spot_list = not_capital['지점명']
words = []

for spot in spot_list:
    if type(spot) != str:
        continue
    
    tmp = kkma.morphs(spot) # 형태소로 분석
    
    for t in tmp:
        words.append(t)

len(words)
```

```python
r_number = re.compile('[0-9]+') # 숫자 제거

words = [r_number.sub('', i) for i in words]
len(words)
```

```python
words = [i for i in words if len(i)>1]
len(words)
```

```python
words_ordered = pd.Series(words).value_counts()
words_ordered[:30]
```

```python
# 불용어 리스트 생성
stopwords_korean = ['부근', '번지', '동쪽', '서쪽', '남쪽', '북쪽', '북서', '북동', '남서', '서북', '남동']

# 불용어 리스트를 활용해 제거
words = [i for i in words if i not in stopwords_korean]
words_ordered = pd.Series(words).value_counts()
words_ordered[:30]
```
